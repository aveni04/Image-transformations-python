```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Step 1: Original Shape (Square) ----
points = np.array([
    [0,0],
    [2,0],
    [2,2],
    [0,2],
    [0,0]  
])

# ---- Step 2: Define Affine Transformation Parameters ----
A = np.array([[1, 0.5],   # shear + scale
              [0.2, 1]])  # shear + scale
t = np.array([1, 2])      # translation

# ---- Step 3: Apply Affine Transformation ----
transformed_points = (A @ points.T).T + t

# ---- Step 4: Plot Both Shapes ----
plt.figure(figsize=(7,7))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid(True, linestyle="--", alpha=0.6)

# Original shape (blue square)
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Square")

# Transformed shape (red quadrilateral)
plt.plot(transformed_points[:,0], transformed_points[:,1], 'ro-', label="Affine Transformed")

plt.legend()
plt.title("Affine Transformation of a Square")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```
