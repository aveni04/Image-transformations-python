```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0] 
])

# ---- Define Euclidean Transformation ----
theta = np.radians(45)   # rotation angle (45 degrees)
R = np.array([[np.cos(theta), -np.sin(theta)],   # rotation matrix
              [np.sin(theta),  np.cos(theta)]])
t = np.array([2, 1])     # translation vector

# Apply transformation: p' = R * p + t
transformed_points = (R @ points.T).T + t


plt.figure(figsize=(6,6))
plt.axhline(0, color='gray', lw=1)  # x-axis
plt.axvline(0, color='gray', lw=1)  # y-axis
plt.grid(True, linestyle="--", alpha=0.6)

# Original shape (blue)
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Shape")

# Transformed shape (red)
plt.plot(transformed_points[:,0], transformed_points[:,1], 'ro-', label="After Euclidean Transformation")

plt.legend()
plt.title("Euclidean Transformation (Rotation + Translation)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```
