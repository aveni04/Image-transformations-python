```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Step 1: Original Shape (Square) ----
points = np.array([
    [0,0,1],
    [2,0,1],
    [2,2,1],
    [0,2,1],
    [0,0,1]  
])

# ---- Step 2: Define Projective Transformation Matrix (Homography) ----
H = np.array([
    [1, 0.3, 1],   # projective skewing
    [0.2, 1, 1], 
    [0.001, 0.001, 1]  # perspective effect
])

# ---- Step 3: Apply Transformation ----
transformed_points = (H @ points.T).T
# Normalize by last coordinate (since projective is up to scale)
transformed_points /= transformed_points[:,2][:,np.newaxis]

# ---- Step 4: Plot Both Shapes ----
plt.figure(figsize=(7,7))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid(True, linestyle="--", alpha=0.6)

# Original square
plt.plot(points[:,0], points[:,1], 'bo-', label="Original Square")

# Projective transformed quadrilateral
plt.plot(transformed_points[:,0], transformed_points[:,1], 'ro-', label="Projective Transformed")

plt.legend()
plt.title("Projective Transformation (Homography)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```
