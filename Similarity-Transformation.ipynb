```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Step 1: Original Shape (Square) ----
points = np.array([
    [0,0],
    [1,0],
    [1,1],
    [0,1],
    [0,0]  
])

# ---- Step 2: Define Similarity Transformation ----
scale = 1.5   # scaling factor (enlarges the square 1.5x)
theta = np.radians(45)   # rotation angle
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])  # rotation matrix
t = np.array([2, 1])     # translation vector (shift right 2, up 1)

# ---- Step 3: Apply Transformation ----
# First scale and rotate, then translate
transformed_points = scale * (R @ points.T).T + t



plt.figure(figsize=(6,6))
plt.axhline(0, color='gray', lw=1)   # x-axis
plt.axvline(0, color='gray', lw=1)   # y-axis
plt.grid(True, linestyle="--", alpha=0.6)

# Original shape (blue)
plt.plot(points[:,0], points[:,1], 'bo-', label="Original")

# Transformed shape (red)
plt.plot(transformed_points[:,0], transformed_points[:,1], 'ro-', label="Similarity Transformed")

plt.legend()
plt.title("Similarity Transformation (Scaling + Rotation + Translation)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis("equal")
plt.show()
```
